4. Задача о раскрое

Сталепрокатное предприятие выпускает стандартные стальные листы размерами 1500×6000 мм. Ежемесячный объем заказов на листы нестандартных размеров приведён ниже: 500×600 — 3000, 1000×400 — 2000, 1500×1000 — 1500 Необходимо составить ЦФ для нахождения наименее затратного раскроя листов.


Пусть у нас имеется m вариантов раскроя, т.е. известны величины известны величины $a_{ij} (i= \overline{1,m}, j=\overline{1,3})$, определяющие количество единиц j-x изделий при i-м способе раскроя единицы сырьевого материала и остаток трубы $a_i$.

Пусть $x_i$ - количество листов, раскраиваемые i-м вариантом $(i=\overline{1,m})$

Тогда количество изделий 1-го вида равно:

$$a_{11}x_1+...+a_{i1}x_i+...+a_{m1}x_m$$

Учитывая условия количества 

$$a_{11}x_1+...+a_{i1}x_i+...+a_{m1}x_m = 3000$$


Аналогичные равенства можно записать для остальных размеров:

$$a_{12}x_1+...+a_{i2}x_i+...+a_{m2}x_m = 2000$$

$$a_{13}x_1+...+a_{i3}x_i+...+a_{m3}x_m = 1500$$



Целевая функция же будет:
$$\sum_{i=1}^{m}x_ia_i \rightarrow min$$


Итого:
$$\sum_{i=1}^{m}x_ia_i \rightarrow min$$

$$a_{1j}x_1+...+a_{ij}x_i+...+a_{mj}x_m = k_j$$

где $k_j$ - необходимое определенных листов

5. Задача о курьерах

Пусть есть компания, занимающаяся доставкой Belivery Club.
Есть N зон, в которых есть спрос на курьеров в каждый час. Внутри дня в разные часы спрос различен. Каждой зоне i в час h необходимо необходимо $n_{hi}$ курьеров. Для каждой зоны есть штраф из-за недостататка 1 курьера - $s_i$

Необходимо распределить M курьеров таким образом, чтобы суммарный штраф был минимальный.
Решить задачу про курьеров, при наличии предпочтей относительно зон и часов у курьеров (описать формулу в общем случае)

Пусть: 
zones = [[],[]...] - список предпочтений курьеров по зонам;
times = [[],[]...] - список предпочтений курьеров по часам

Пронумеруем курьеров:

пусть $x_{ih}^k$ = 0, если (зона i не входит в zones[k]) ИЛИ (час h не входит в times[k])

пусть $x_{ih}^k$ = 1, если k-курьер работает в час h в i зоне

ELSE $x_{ih}^k$ = 0


Тогда в одной зоне в час h работает $$\sum_{k=1}^{M} x_{ih}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} x_{ih}^k - n_{ih} >0 \\
s_i * (\sum_{k=1}^{M} x_{ih}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 



Штраф за день $$\sum_{h=1}^{24}f_{ih}$$

Штраф по всем зонам в течении дня:


$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}$$

Целевая функция:

$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih} \rightarrow min $$
